````python
import csv
import matplotlib.pyplot as plt
import pandas as pd
from numpy import mean

cars = open('C:\\Users\\USer\\Desktop\\python\\resource\\cars.csv','r')
df = pd.read_csv('C:\\Users\\USer\\Desktop\\python\\resource\\cars.csv', sep = '\t')

speed = df['speed'] #x
dist = df['dist'] #y
n= len(df)

Sxy = sum(speed*dist)*n - sum(speed)*sum(dist)
Sxx = sum(speed^2)*n - sum(speed)^2

b1 = Sxy/Sxx
b0 = mean(dist) - mean(speed)*b1


plt.scatter(df['speed'],df['dist'])
plt.xlabel('speed')
plt.ylabel('dist')
plt.plot(speed, b1*speed+b0)
plt.show()
````
